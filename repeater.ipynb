```python
import pyautogui
import time
import subprocess
import os
def savePartTwo():
    def saveRepeaterLog():
        repetitions = 100
        start_number = 1
        time.sleep(5)
        #Go to repeater
        size=0
        pyautogui.hotkey('ctrl', 'shift', 'r')
        for i in range(repetitions):
            
            # Step 1: Save history
            pyautogui.hotkey('ctrl', 'j')
            time.sleep(.5)  
            # Step 2: Type "repeater{number}"
            pyautogui.write(f'repeater{start_number + i}', interval=0.1)
            time.sleep(.5)
            # Step 3: Press Enter twice to save the file
            pyautogui.press('enter')
            time.sleep(.6)
          
            pyautogui.press('enter')
            time.sleep(.6)
            curr_size = os.path.getsize(rf'D:\burpsuite automation\myfiles{start_number + i}')
            print("size of :",f'repeater{start_number + i}',"is ",curr_size)
            if(i==0):
                size=curr_size
            elif(curr_size==size):
                break
            # step 4:Add a delay before repeating the loop
            
            # Step 5: Press ctrl+'+' to go to next tab
            pyautogui.hotkey('ctrl', '=')
            time.sleep(.6)
            
    def runScripts():
        subprocess.run(['python', 'createXML.py'], cwd='./myfiles/')
        subprocess.run(['python', 'createCSV.py'], cwd='./myfiles/')
        subprocess.run(['python', 'delete.py'], cwd='./myfiles/')
        subprocess.run(['python', 'cut.py'], cwd='./myfiles/')
        subprocess.run(['python', 'merge.py'], cwd='./myfiles/')

    saveRepeaterLog()
    runScripts()

```
